```python
#info sur un noeud
#pour un pas et un fenetre
step = 1000
Dasn = 20000
asn_fin = 0
asn_debut = 0
addr_1 = "8772"
taux_transm = []
list_neig = []
nb_neig = 0
#liste des msg envoyés entre noeud1 et 2 | qui appartienent a: [asn_data-Dasn; asn_data]
list_ev_tx = []
del list_ev_tx[:]

with open("data/parsed/event/data_STAT_DATAGEN.log", "r") as f_tx:
    for line in f_tx:
        pass
    asn_fin = int(get_asn(line))

#pour chaque iteration, on regarde quels sont les voisins | et on calcule le taux de transmission
#pour chacun d'entre eux.
while (asn_debut < asn_fin):
    del taux_transm[:]
    taux_transm = []
    nb_neig, list_neig = get_neighbor_node(addr_1,asn_debut)
    if nb_neig != 0:
        for neighbor in list_neig:

            ################################################################################
            asn_data=asn_debut
            addr_2= neighbor
            PT_OF_VIEW_NODE=False
            #liste des msg envoyés entre noeud1 et 2 | qui appartienent a: [asn_data-Dasn; asn_data]
            list_ev_tx = []
            del list_ev_tx[:]
            #contient 0 ou 1 si le msg est recu
            list_recu_tx = [] 
            del list_recu_tx[:]
            #resultat
            taux_transm_n = -1

            #on recupere les envois
            with open("data/parsed/event/data_STAT_PK_TX.log", "r") as f_tx:
                for line in f_tx:
                    if (addr_1 == get_addr(line) and get_l2Dest(line).endswith(addr_2)
                       and int(get_asn(line)) > (asn_data-Dasn) and int(get_asn(line)) < asn_data):
                        list_ev_tx.append(make_event_PK_TX(get_time(line), 
                                                                   get_addr(line), 
                                                                   get_comp(line), 
                                                                   get_asn(line), 
                                                                   get_statType(line), 
                                                                   get_trackinstance(line), 
                                                                   get_trackowner(line), 
                                                                   get_length(line), 
                                                                   get_frameType(line), 
                                                                   get_slotOffset(line), 
                                                                   get_frequency(line), 
                                                                   get_l2Dest(line), 
                                                                   get_txpower(line), 
                                                                   get_numTxAttempts(line),
                                                                   get_queuePos(line),
                                                                   0.))


            if list_ev_tx:
                if PT_OF_VIEW_NODE == True:
                    #on verifie si ces msgs ont été recus | VUE PAR LE NOEUD
                    with open("data/parsed/event/data_STAT_ACK_RX.log", "r") as f_rx:
                        for line in f_rx:
                            for trans in list_ev_tx:
                                #si le noeud a recu un ack a cet asn
                                if (get_addr(line) == addr_1 and get_asn(line) == trans.asn):
                                    trans.recu=1.
                                    
                                #on sort de la boucle si on a depasser l'asn pour economiser du temps
                            #if (get_asn(line) > list_ev_tx[-1].asn):
                                #break
                else:
                    #on verifie si ces msgs ont été recus | VUE PAR LA RECEPTION DU MSG
                    with open("data/parsed/event/data_STAT_PK_RX.log", "r") as f_rx:
                        for line in f_rx:
                            #on sort de la boucle si on a depasser l'asn pour economiser du temps  
                            if (int(get_asn(line)) > int(list_ev_tx[-1].asn)):
                                break
                            for trans in list_ev_tx:
                                #si le dest a recu un msg a ce moment
                                if (get_addr(line) == addr_2 and get_l2Src(line).endswith(addr_1) and 
                                    get_asn(line) == trans.asn):
                                    trans.recu=1.

            #on calcule le taux
            if list_ev_tx:                
                for elm in list_ev_tx:
                    list_recu_tx.append(elm.recu)

                taux_transm_n = np.sum(list_recu_tx)/len(list_recu_tx)

            taux_transm.append(taux_transm_n)

    #if list_ev_tx:
        #print list_ev_tx, list_ev_tx[-1].asn
    #for el in list_ev_tx:
        #print el.recu
    ################################################################################        
        
    ### affichage    
    print "asn : ", asn_debut, "| nb de voisins", nb_neig
    if nb_neig != 0:
        print "    voisins : ", list_neig
        for neighbor in list_neig:
            print "    voisin : ", neighbor, "taux de transm : ", taux_transm[list_neig.index(neighbor)]
        
        
        
    asn_debut = asn_debut + step

print 1
```
